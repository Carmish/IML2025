Given the affine model: 
\[y = w_1 x_1 + w_2 x_2 + w_3\]
Determine which of the weight options $(w_1, w_2, w_3)$ contain all data points below, such that all three points are on the plane:
**Data points:**
Input $x_i$, label $y$:
- $x_1 = 2, x_2 = 0, y = 5  $
- $x_1 = 1, x_2 = 1, y= 6  $
- $x_1 = 0, x_2 = 2, y= 7  $

**Options:**
1. $(w_1, w_2, w_3) = (1, 0, 3)$
2. $(w_1, w_2, w_3) = (1, 3, 2)$
3. $(w_1, w_2, w_3) = (2, 3, 1)$



</div></article>



```python
#Write your answer here...
def f(w1, w2, w3):
    if ((0 == (w1*2+w2*0+w3-5)) and (0 == (w1*1+w2*1+w3-6)) and (0 == (w1*0+w2*2+w3-7))):
        return f"Weight option {w1, w2, w3} contains all datapoints"
    else:
        return f"Weight option {w1, w2, w3} does not contain all datapoints"


print(f(1, 0, 3))
print(f(1, 3, 2))
print(f(2, 3, 1))

# Weight option (1, 0, 3) does not contain all datapoints
# Weight option (1, 3, 2) does not contain all datapoints
# Weight option (2, 3, 1) contains all datapoints
```

### The notation for rewriting the affine model as an **inner product**
_(Obs not from Dan, so be aware of possible mistakes)_

Given the affine model:
$[ y = w_1 x_1 + w_2 x_2 + w_3 ]$

You can express this as an inner product by augmenting the input vector with a 1 (to account for the bias term $( w_3 )$):
$$[
y = 
\begin{bmatrix}
w_1 \\
w_2 \\
w_3
\end{bmatrix}^T
\begin{bmatrix}
x_1 \\
x_2 \\
1
\end{bmatrix}
= \langle w, x \rangle
]$$

Here, $( w = [w_1, w_2, w_3]^T )$ and $( x = [x_1, x_2, 1]^T )$.

So, for each data point, the model is:
$[
y_i = \langle w, x_i \rangle
]$
where $( x_i = [x_{1}, x_{2}, 1]^T )$.

This is the standard way to write an affine model as an inner product.
(the inner product is between the weight vector and the augmented input vector)

### The notation for rewriting the affine model in **matrix form** 
_(Obs not from Dan, so be aware of possible mistakes)_

Given the model:
$$[
y = w_1 x_1 + w_2 x_2 + w_3
]$$

For multiple data points, you can write this as:
$$[
Aw = y
]$$
where

$$[
A = \begin{bmatrix}
x_{1,1} & x_{1,2} & 1 \\
x_{2,1} & x_{2,2} & 1 \\
x_{3,1} & x_{3,2} & 1 \\
\end{bmatrix}
,\quad
w = \begin{bmatrix}
w_1 \\
w_2 \\
w_3
\end{bmatrix}
,\quad
y = \begin{bmatrix}
y_1 \\
y_2 \\
y_3
\end{bmatrix}
]$$

For your specific example:
$$[
A = \begin{bmatrix}
1 & 0 & 1 \\
0 & 1 & 1 \\
1 & 1 & 1 \\
\end{bmatrix}
,\quad
w = \begin{bmatrix}
w_1 \\
w_2 \\
w_3
\end{bmatrix}
,\quad
y = \begin{bmatrix}
2 \\
3 \\
6
\end{bmatrix}
]$$

So the system is:

$$[
\begin{bmatrix}
1 & 0 & 1 \\
0 & 1 & 1 \\
1 & 1 & 1 \\
\end{bmatrix}
\begin{bmatrix}
w_1 \\
w_2 \\
w_3
\end{bmatrix} 
= 
\begin{bmatrix}
2 \\
3 \\
6
\end{bmatrix}
]$$

This is the standard matrix notation for the affine model with multiple data points.

**Note to self on how to solve:**

For each case, check if the model is linear in the **inputs**. That is, can the model be written in the form:
$$
y = a_1 x_1 + a_2 x_2 + \ldots + a_n x_n + b
$$
where each $x_i$ is an input variable (not a function of the inputs), and $a_i$, $b$ are constants (possibly weights).

---

**A:** $y = w_1 \cos(x_1) + w_2 x_2^2 + w_3$  
- **Not linear in the inputs**: $\cos(x_1)$ and $x_2^2$ are nonlinear functions of the inputs.

**B:** $y = e^{x_1} w_1 + w_2 x_2^2 + w_3$  
- **Not linear in the inputs**: $e^{x_1}$ and $x_2^2$ are nonlinear in the inputs.

**C:** $y = w_3 + \sin(x_1) w_1 + w_2 \ln(x_2)$  
- **Not linear in the inputs**: $\sin(x_1)$ and $\ln(x_2)$ are nonlinear in the inputs.

**D:** $y = w_1 x_1 x_2 + w_2 x_2 + w_3$  
- **Not linear in the inputs**: $x_1 x_2$ is a product of inputs (nonlinear).

**E:** $y = w_1 x_1 + w_2 \sin(x_2) + w_3 x_3 + w_4 \cos(x_4) + w_5$  
- **Not linear in the inputs**: $\sin(x_2)$ and $\cos(x_4)$ are nonlinear in the inputs.

**F:** $y = w_1 x_1 + w_4 \sin(x_2) + w_2 x_3 + w_3 \cos(x_4) + w_4$  
- **Not linear in the inputs**: $\sin(x_2)$ and $\cos(x_4)$ are nonlinear in the inputs.

---

**Summary:**  
None of the models A–F are linear in the inputs, because each contains at least one nonlinear function of the input variables (such as $\sin$, $\cos$, $\ln$, $e^{x_1}$, or a product like $x_1 x_2$).  
A model is only linear in the inputs if each term is a constant times a single input variable (or a constant), with no nonlinear transformations or products of input variables.

1. How do these exercises relate to fitting a model to data?

These exercises show how to find parameters (weights) for a model that fits given data points. By solving equations based on input-output pairs, we can determine the weights that minimize the error between the model's predictions and the actual labels. This is what we do in machine learning where we try to determine a function from data.

2. Each of the exercises above are designed to have a single exact solution. How does assuming a single exact solution relate to real world problems? Consider inputs and labels as part of your argument. 

In real-world problems, data often contains noise and uncertainty, making it unlikely to have a single exact solution. 
In practice, we often seek approximate solutions that generalize well to unseen data rather than exact fits. We see this e.g. in regression tasks where we minimize a loss function (like mean squared error) rather than solving for exact weights.
Considering inputs and labels, real-world data may have measurement errors, which means that multiple models could fit the data reasonably well. So instead of a single exact solution, we often look for a best-fit solution that balances accuracy and generalization, so the function should be satisfied even if the data is a bit off.

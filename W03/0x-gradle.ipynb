# Visualizing linear transformations

```python
#QUESTION #21371

def T(x): 
    #Conclusion: T rotates all vectors over pi radians (180 degrees) around the origin.
    return ((np.array([[-1, 0], 
                       [0, -1]])) @ x )

u = np.array([3,-3])
v = np.array([-3,-3])
Tu = T(u)
Tv = T(v)

# plt.arrow(0,0,v[0],v[1], color='blue')
# plt.arrow(0,0,u[0],u[1], color='red')
# plt.arrow(0,0,Tv[0],Tv[1], color='purple')
# plt.arrow(0,0,Tu[0],Tu[1], color='pink')
# plt.show


# Plot setup
plt.figure(figsize=(6,6))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlim(-4, 4)
plt.ylim(-4, 4)

# Original vectors
plt.arrow(0, 0, u[0], u[1], head_width=0.1, color='red', length_includes_head=True, label='u')
plt.arrow(0, 0, v[0], v[1], head_width=0.1, color='blue', length_includes_head=True, label='v')

# Transformed vectors
plt.arrow(0, 0, Tu[0], Tu[1], head_width=0.1, color='pink', length_includes_head=True, label='T(u)')
plt.arrow(0, 0, Tv[0], Tv[1], head_width=0.1, color='purple', length_includes_head=True, label='T(v)')

#plt.legend(["u", "v", "T(u)", "T(v)"])
plt.legend(loc=3,bbox_to_anchor=(1,0))
plt.title("Original and Transformed Vectors")
plt.show()
```

```python
from fractions import Fraction
import numpy as np

def frac(decimal_matrix):
    ### Formats all non-int elements as fractions:

    fraction_matrix = np.zeros_like(decimal_matrix, dtype=object)

    for i in range(decimal_matrix.shape[0]):
        for j in range(decimal_matrix.shape[1]):
            if (decimal_matrix[i, j] % 1 == 0):
                fraction_matrix[i, j] = int(decimal_matrix[i, j])
            else:
                fraction_matrix[i, j] = str(Fraction(decimal_matrix[i, j]).limit_denominator())
    
    return fraction_matrix

def solve22(a, b, c, d): 
    ### Finds the determinant and inverse of a 2x2 matrix
    A = np.array([[a,b],[c,d]])
    print("\nA: ", A)
    detA = np.linalg.det(A)
    print(" Det: ", detA)
    if (detA==0):
        print(" Not invertable")
    else: 
        print(" Inv: ", frac(np.linalg.inv(A)))

def solve33(a, b, c, d, e, f, g, h, i): 
    ### Finds the determinant and inverse of a 3x3 matrix
    A = np.array([[a,b,c],[d,e,f],[g,h,i]])
    print("\nA: ", A)
    detA = np.linalg.det(A)
    print(" Det: ", detA)
    if (detA==0):
        print(" Not invertable")
    else: 
        print(" Inv: ", frac(np.linalg.inv(A)))



#Exercises
solve22(0, 1, 5, 0)
solve22(4, 3, 0, 0)
solve22(3, 2, -5, 2)
solve22(1, -5, -4, -3)
solve22(-5, -2, -3, 3)
solve22(-2, 1, -2, 1)
solve22(-2, -1, 2, 1)
solve22(4, 5, -3, 3)
solve22(5, 4, -6, 1)
solve22(2, 3, -2, 5)
solve33(-5, 3, 0, -3, -4, -5, 4, 4, 5)
solve33(1, 3, 0, -3, -8, -4, -1, -1, -6)
solve33(1, -2, 0, -1, 3, 2, 0, 0, 2)

solve33(2, -3, 3, 2, -4, 6, 0, -3, 9)
```

```python
def solve_for_X(a, b):
    # Determine X s.t. aX = b when a is invertible
    A_inv = np.linalg.inv(a)
    X = A_inv @ b
    return X

#QUESTION #21476
a = np.array([[1, -3, 1], [3, -8, 4], [-3, 10, -1]])
b = np.array([[-3,0,-1], [-2, 3, 3], [1, 0, 0]])
X = solve_for_X(a, b)
print("Matrix X (QUESTION #21476):\n", X)

#QUESTION #21477
a = np.array([[1, 2, 3], [0, 1, 1], [1, -1, 1]])
b = np.array([[2, -3, 1], [2, 0, -1], [1, 3, -3]])
X = solve_for_X(a, b)
print("Matrix X (QUESTION #21477):\n", X)
```
